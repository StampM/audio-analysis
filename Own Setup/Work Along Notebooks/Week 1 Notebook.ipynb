```python
from scipy.io.wavfile import read
%matplotlib inline
```

```python
help(read)
# shows input and output - reutrns sampling rate and array with sound samples

```

```python


(fs, x) = read('flute-A4.wav')

x
# x - array of 16-bit integers 
```

```python
len(x) # how many sample
len(x) / fs # seconds ran
```

```python
import matplotlib.pyplot as plt
plt.plot(x)
# visualsie but not in best way
```

```python
# plot in seconds

import numpy as np
t = np.arange(x.size)/fs

plt.plot(t, x) # now shwo reconds

# in IPython Pop-up box - we can zoom in section
```

```python
# select samples

y = x[100:1000]
plt.plot(y) # can see zoom in
```

```python
np.max(y) 
# sound oscillates around 0 
np.max(abs(y)) 
# comnsiders negative as considered
```

```python
# can write output

from scipy.io.wavfile import write
write('test.wav', fs, y)
# need fs first for sampling rate
```
