```python
import numpy as np
from scipy.signal import get_window
from scipy.fftpack import fft
import math 
import matplotlib.pyplot as plt
```

```python
# get length
M = 63
# get window - hanning is raised cosine
window = get_window('hanning', M) 
# vars to keep middle
hm1 = int(math.floor((M+1)/2))
hm2 = int(math.floor(M/2))
```

```python
# prepare window for FFT comptuing
N = 512 # FFT size
hN = N//2 # says N/2 + 1 but would not work
# place window around 0th sample
fftbuffer = np.zeros(N)
fftbuffer[:hm1] = window[hm2:]
fftbuffer[N-hm2:] = window[:hm2]
```

```python
# calculate fft spectrum and change to decibels

X = fft(fftbuffer)

absX = abs(X)

absX[absX < np.finfo(float).eps] = np.finfo(float).eps
#. eps - min value to have in python - make sure that log is not error
mX = 20 * np.log10(absX)
pX = np.angle(X) 
```

```python
mX.size
```

```python
# undow zero-phase window to place in middle
mX1 = np.zeros(N)
pX1 = np.zeros(N)

mX1[:hN] = mX[hN:]
mX1[N-hN:] = mX[:hN]

pX1[:hN] = pX[hN:]
pX1[N-hN:] = pX[:hN]
```
